```python
plt.figure(figsize=(10, 7))
x = [3, 4, 5, 6, 7] # данные приблезительные
y1 = [380, 400, 405, 375, 400]
y2 = [375, 335, 380, 365, 360]
y3 = [390, 345, 365, 390, 390]
y4 = [365, 375, 390, 370, 385]
plt.plot(x, y1, c='green')
plt.plot(x, y2, c='blue')
plt.plot(x, y2, c='orange')
plt.plot(x, y2, c='red')
plt.legend(['Open', 'High', 'Low', 'Close'])
plt.plot(x, y1, '-', x, y2, '-', x, y3, '-', x, y4, '-')
```

```python
import datetime as DT
from matplotlib import pyplot as plt
from matplotlib.dates import date2num
data = [(DT.datetime.strptime('2016-10-03', "%Y-%m-%d"), 772.5),
        (DT.datetime.strptime('2016-10-04', "%Y-%m-%d"), 776.3),
        (DT.datetime.strptime('2016-10-05', "%Y-%m-%d"), 776.5),
        (DT.datetime.strptime('2016-10-06', "%Y-%m-%d"), 777),
        (DT.datetime.strptime('2016-10-07', "%Y-%m-%d"), 775.2)]
x = [date2num(date) for (date, value) in data]
y = [value for (date, value) in data]
fig = plt.figure()
graph = fig.add_subplot(111)
graph.plot(x,y,'r-o')
graph.set_xticks(x)
graph.set_xticklabels(
        [date.strftime("%Y-%m-%d") for (date, value) in data])
plt.xlabel('Date')
plt.ylabel('Closing Value')
plt.title('Closing stock value of Alphabet Corparated', color='black') 
plt.minorticks_on()
plt.grid(which='major', linestyle='--', linewidth='0.9', color='blue')
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.1, 17.5, 8.9, 8, 7.8, 6.8] # приблезительные значения
x_pos = [i for i, _ in enumerate(x)]
plt.bar(x_pos, popularity, color='dodgerblue')
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.title("PopularitY of Programming Language\n" + "Worldwide, Oct 2017 compared to a year ago")
plt.xticks(x_pos, x)
# основная сетка
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')
# дополнительная сетка
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```
